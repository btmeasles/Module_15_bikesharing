```python
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
Citibike_df = pd.read_csv('NYC_CitiBike_Original_StarterCode/201908-citibike-tripdata.csv')
Citibike_df.head()
```

```python
# 2. Check the datatypes of your columns. 
Citibike_df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
Citibike_df['tripduration'] = pd.to_datetime(Citibike_df['tripduration'], unit='s')

Citibike_df['Trip Duration'] = Citibike_df['tripduration'].dt.strftime('%H:%M:%S')
```

```python
# 4. Check the datatypes of your columns. 
Citibike_df.dtypes
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
# ----------------
# Some Other stuff performed first ↓↓↓
```

```python
Citibike_df.starttime = pd.to_datetime(Citibike_df.starttime)
Citibike_df.stoptime = pd.to_datetime(Citibike_df.stoptime)
```

```python
Citibike_df['Age'] = 2019 - Citibike_df['birth year']
Citibike_df['Age'].max()
# Say what!? 
```

```python
Citibike_df = Citibike_df.rename(columns={'starttime':'Start Time','stoptime':'Stop Time','start station id':'Start Station ID', 'start station name':'Start Station Name','start station latitude':'Start Station Latitude','start station longitude':'Start Station Longitude','end station id':'End Station ID','end station name':'End Station Name','end station latitude':'End Station Latitude','end station longitude':'End Station Longitude','bikeid':'Bike ID', 'usertype':'User Type','gender':'Gender'})
Citibike_df.head()
```

```python
Citibike_df = Citibike_df.drop(columns=['tripduration', 'birth year'])
Citibike_df.head()
```

```python
Citibike_df = Citibike_df[['Trip Duration','Start Time',
 'Stop Time',
 'Start Station ID',
 'Start Station Name',
 'Start Station Latitude',
 'Start Station Longitude',
 'End Station ID',
 'End Station Name',
 'End Station Latitude',
 'End Station Longitude',
 'Bike ID',
 'User Type',
 'Gender',
 'Age']]
Citibike_df.head()
```

```python
Citibike_df.to_csv('NYC_CitiBike_Challenge.csv')
```
